```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/titanic_train.csv'
df = pd.read_csv(csv)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
survived = df.PassengerId[df.Survived == 1].count()
died = df.PassengerId[df.Survived == 0].count()
```

```python
labels = ['Survived', 'Died']
colours = ['Red', 'Yellow']
plt.figure(figsize=(8,5))

plt.pie([survived, died], labels = labels, colors=colours, autopct='%.2f %%')

plt.title('Survivors vs Casualities of Titanic Disaster')
plt.show()
```

```python
male = df.PassengerId[df.Sex == 'male'].count()
female = df.PassengerId[df.Sex == 'female'].count()
```

```python
labels = ['Male', 'Female']
colours = ['Red', 'Yellow']
plt.figure(figsize=(8,5))

plt.pie([male, female], labels = labels, colors=colours, autopct='%.2f %%')

plt.title('Males vs Females of Titanic Disaster')
plt.show()
```
