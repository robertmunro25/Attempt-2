```python
y = [1,4,9,16,25,36,49,64,81]
x = [1,2,3,4,5,6,7,8,9]

plt.figure(figsize = (4,2), dpi=100)

x2 = np.arange(0,10,1)
plt.plot(x2, x2**3)

plt.plot(x,y, label='X Squared', color='R', linewidth=1, marker = 'x', linestyle=':')

plt.title('Y = X Squared')
plt.xlabel('X')
plt.ylabel('Y')

plt.savefig('mygraph.png')
```
