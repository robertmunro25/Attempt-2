### **Cleaning Up the Data!!!!**

```python
nandf = year[year.isna().any(axis=1)]
nandf.head()

year = year.dropna(how='all')
```

```python
year['Quantity Ordered'] = year['Quantity Ordered'].astype('int32')
year['Price Each'] = year['Price Each'].astype('float_')
```

### **Adding a Sales Column**

```python
year['Sales'] = year['Quantity Ordered'] * year['Price Each']
```

### **What City Had Highest Number of Sales**

```python
year
```

```python
def get_city(address):
  return address.split(',')[1]

def get_state(address):
  return address.split(',')[2].split(' ')[1]

year['City'] = year['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
```

```python
year
```

```python
best_city = year.groupby(['City']).sum()

best_city
```

```python
cities = [city for city, df in year.groupby('City')]

plt.bar(cities, best_city.Sales)

plt.xticks(cities, rotation='vertical')
plt.xlabel('Cities')
plt.ylabel('Total Sales in Millions of USD')

plt.show()
```

### **What Time Should We Display Advertisements to Maximise Sales?**

**What Objects are Most Often Sold Together?**

```python
orders = year[year['Order ID'].duplicated(keep=False)]

orders['Grouped'] = orders.groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))

orders = orders[['Order ID', 'Grouped']].drop_duplicates()

orders
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in orders['Grouped']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list, 2)))

data = count.most_common(10)
```
