```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/us-weather-history/KSEA.csv'
```

```python
df = pd.read_csv(csv)
```

```python
df.head()
```

```python
df.info()
```

```python
df.date = pd.to_datetime(df.date)
```

```python
df.info()
```

```python
df.head()
```

```python
mean = df.actual_mean_temp
max = df.actual_max_temp
min = df.actual_min_temp

plt.plot(df.date, mean, label='Mean Temp.', color='R')
plt.plot(df.date, max, label='Max Temp.', color='B')
plt.plot(df.date, min, label='Min Temp.', color='G')

plt.xlabel('Date')
plt.ylabel('Temp. (F)')
plt.legend()
```

```python
months = df.groupby(pd.Grouper(key='date', axis=0, freq='M')).mean()
```

```python
months
```

```python
m_mean = months.actual_mean_temp
m_max = months.actual_max_temp
m_min = months.actual_min_temp

dates = ['07', '08', '09', '10', '11', '12', '01', '02', '03', '04', '05', '06']

plt.plot(dates, m_mean, 'y.-', label='Mean Temp.')
plt.plot(dates, m_max, 'r.-', label='Max Temp.')
plt.plot(dates, m_min, 'b.-', label='Min Temp.')

plt.xlabel('Month')
plt.ylabel('Temp. (F)')

plt.legend()
```
